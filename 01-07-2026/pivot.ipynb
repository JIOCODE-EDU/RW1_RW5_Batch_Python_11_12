### Practical Tasks for Pivoting, Re-Indexing, Groupby and Transform in Pandas

```python
import pandas as pd

print("=" * 40)
print("1. Pivoting DataFrames")
print("=" * 40)
```

#### Monthly Sales Pivot

```python
sales = pd.DataFrame({
  "Month":["Jan" , "Jan" ,  "Feb" , "Feb",  "Mar" , "Mar"],
  "Product":["Laptop" , "Mobile" , "Laptop" , "Mobile" , "Laptop" , "Mobile"],
  "Sales":[50000 , 30000 , 55000 , 35000 , 60000 , 40000]
})

# print(sales)

pivot_sales = sales.pivot(index="Month" , columns="Product" , values="Sales")

month_order = ["Jan" , "Feb" , "Mar"]

pivot_sales = pivot_sales.reindex(month_order)

print(pivot_sales)
```

#### Student Marks Pivot

```python
marks = pd.DataFrame({
  "Student":["Amit", "Amit" , "Raj" , "Raj" , "Neha",  "Neha"],
  "Subject":["Science" , "Math" , "Science" , "Math" , "Science" , "Math"],
  "Marks":[80 , 78 , 90 , 88 , 95 , 92]
})

print(marks)
```

```python
pivot_marks = marks.pivot(index="Student" , columns="Subject" , values="Marks")

print(pivot_marks)
```
