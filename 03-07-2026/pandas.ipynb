```python
import pandas as pd
```

```python
data = pd.DataFrame({
  "ID":[101 , 102 , 103 , 104 , 105 , 106 , 107 , 108],
  "Name":["Amit" , "Pooja" , "Arti" , "Priya" , "Neha" , "Rohan",  "Rahul" , "Sneha"],
  "Gender":["Male" , "Female" , "Female" , "Female" , "Female" , "Male" , "Male" , "Female"],
  "Department":["IT" , "HR" , "Finance" , "IT" , "HR" , "Finance" , "HR" , "IT"],
  "Performance":["Good" , "Excellent" , "Average" ,  "Good" , "Average",  "Average",  "Excellent" , "Excellent"]
})

print(data)
```

#### Handle Categorical Variables

###### Convert Object into Category

```python
data["Gender"] = data["Gender"].astype("category")
data["Department"] = data["Department"].astype("category")
data["Performance"] = data["Performance"].astype("category")

print(data.dtypes)

print(data["Gender"].cat.categories)

print(data["Department"].cat.categories)

print(data["Performance"].cat.categories)

print(data["Gender"].cat.codes)

print(data["Department"].cat.codes)
```

#### Perform Categorical Analysis

```python
print(data["Department"].value_counts())
print(data["Gender"].value_counts())
print(data["Performance"].value_counts())
print(data["Department"].value_counts(normalize = True) * 100)
print(data["Gender"].value_counts(normalize = True) * 100)
print(data["Performance"].value_counts(normalize = True) * 100)
print(data["Department"].unique())
print(data["Department"].nunique())
print(data.nunique())
```

#### Compare and order categories

```python
data["Performance"] = data["Performance"].cat.reorder_categories(['Good' , 'Average' , 'Excellent'] , ordered=True)

print(data["Performance"].cat.categories)

print(data.sort_values("Performance"))

print(data[data["Performance"] == "Good"])
```

#### Rename Categories

```python
data["Department"] = data["Department"].cat.rename_categories({
  "IT"  : "Information Technology",
  "HR"  : "Human Resource",
  "Finance" : "Accounts"
})

print(data["Department"])
```

```python
print(data)
```

#### Add new Categories

```python
data["Department"] = data["Department"].cat.add_categories(['Development'])

print(data["Department"].cat.categories)

data["Department"] = data["Department"].cat.remove_categories(['Sales'])

print(data["Department"].cat.categories)
```

```python
data["Department"] = data["Department"].cat.unused_categories()

print(data["Department"].cat.categories)
```
