# Pandas in Python

```python
import pandas as pd
import numpy as np
```

#### One-Dimensional labeled Array
#### Two-Dimensional labeled Array

```python
fruits = pd.Series(["apple", "Banana" , "cherry" , "date" , "apple"])

print("fruits Series :")
print(fruits)
```

## Creating a series with custom indices

```python
colors = pd.Series(["red" , "green" , "blue"] , index=['color1' , 'color2' , 'color3'])

print('Colors Series:')

print(colors)
```

```python
print('/n Fruit at index 1:')
print(fruits[1])
```

```python
print('Colors with index color2 : ')
print(colors['color2'])
```

#### Creating a DataFrame from a Dictionary of Lists

```python
student_data = pd.DataFrame({
  'Name':['Alice' , 'Bob' , 'Charlie' , 'David'],
  'Age':[21 , 25 , 35 , 29],
  'Major':['CS' , 'Engineering' , 'Math' , 'Physics']
})

print('\n Student Data:')

print(student_data)
```

```python
employee_data = pd.DataFrame([{'name': 'David', 'salary': 50000, 'department': 'Sales'},
{'name': 'Eve', 'salary': 60000, 'department': 'Marketing'},
{'name': 'Frank', 'salary': 55000, 'department': 'Sales'}])

print(employee_data)
```

```python
data_array = np.random.randint(1 , 11 , size = (3 , 2))

random_df = pd.DataFrame(
  data_array , columns=["column A" , "column B"]
)

print(random_df)
```

```python
print("\n Number of Elements:")
print(len(fruits))

print("\n Unique Values:")
print(fruits.unique())

print("\n Count of Each Unique Value:")
print(fruits.value_counts())

print("\n First Two Rows:")
print(student_data.head(2))

print('\n Last row of DataFrame:')
print(student_data.tail(1))

print('\n Index of DataFrame:')
print(student_data.index)

print('\n Column of DataFrame:')
print(student_data.columns)

print('\n Consise Data from DataFrame:')
print(student_data.info())

print('\n Name Column:')
print(employee_data['name'])

print("\n Row at index 1:")
print(employee_data.loc[1])

print('\n Name and Salary Column:')
print(employee_data[['name' , 'salary']])

print('\n First two row (name and Department)')
print(employee_data[['name' , 'department']].loc[2])
print(employee_data.loc[:1 , ['name' , 'department']])
```
