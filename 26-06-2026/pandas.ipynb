## Create , Read and Write CSV , Excel & JSON

```python
import pandas as pd
```

#### Create Dictionary

```python
student_data = {
  "Name":["Vivek" , "Rahul" , "Neha" , "Priya" , "Zeel"],
  "Age":[28 , 25 , 21 , 23 , 50],
  "Marks":[89 , 78, 91 , None , 89],
  "City":["Surat" , "Ahmedabad" , "Vadodara" , None , "Rajkot"],
  "JoiningDate":["15-06-2026" , "20-05-2025",  "20-06-2024" , "01-02-2023", "01-05-2026"]
}

df = pd.DataFrame(student_data)

print(df)
```

#### to_csv

```python
df.to_csv("students.csv" , index = False)
print("\n CSV File Created Successfully!!!")
```

### to_excel

```python
df.to_excel("student.xlsx" , index=False)

print("Excel File Created Successfully!!")
```

JSON

```python
df.to_json("studentData.json" , orient="records" , index = 4)
print("JSON file created sucessfully!!")
```
