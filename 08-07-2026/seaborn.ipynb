```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

##### Load Built-in Dataset

```python
tips = sns.load_dataset('tips')
tips.info()
tips.describe()
tips.shape
tips.columns
tips.isnull().sum()
tips.head()
```

##### Bar Plot

```python
plt.figure(figsize=(8 , 5))
sns.barplot(x = "sex" , y="tip" , data=tips , estimator=np.sum , errorbar=None)
plt.xlabel('sex')
plt.ylabel('tip')
plt.title('Total tips by sex')
plt.show()
```

```python
df = sns.load_dataset('titanic')

df.info()
```

```python
plt.figure(figsize=(8 , 5))
sns.barplot(x = 'class' , y='fare' , data = df , estimator=np.sum , errorbar=None)
plt.xlabel('Passenger Class')
plt.ylabel('Total Fare')
plt.title('Total Fare by Class')
plt.show()
```

##### Count Plot

```python
plt.figure(figsize=(8 , 5))
sns.countplot(
  x = "class",
  data = df
)

plt.xlabel('Passanger Class')
plt.ylabel('Number of Passangers')
plt.title('Passenger Count By Class')
plt.show()
```

##### Histogram

```python
plt.figure(figsize=(8 , 5))
sns.histplot(
  x = 'age',
  data = df,
  bins = 20
)

plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
```
