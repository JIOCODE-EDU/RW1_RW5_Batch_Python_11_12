# Pandas in Python

#### Pandas is an open-source Python Library used for :

1. Data Analysis
2. Data Cleaning
3. Data Manipulation
4. Data Visualization 

### Pandas Structure Types:

1. Series (1 - Dimensional)
2. DataFrame (2 - Dimensional)


# Series in Python

#### A series is a one-dimensional labeled array capable of holding any Data type.

```python
import pandas as pd
```

```python
data = pd.Series([])

print(data)
```

```python
data = pd.Series([1 , 2 , 3 , 4 , 5])

print(data)
```

#### Series with custom Index

```python
data = pd.Series(
  [90 , 85 , 95],
  index = ['Math' , 'Science' , 'English']
)

print(data)
```

#### Access Series Data

```python
print(data['Math'])
```

# DataFrame in Python


#### A DataFrame is a two-dimensional labeld tubular structure considering of rows and columns.

```python
student = {
  "Name":["Rahul" , "Ritu" , "Rajesh" , "Dixit"],
  "Age":[20 , 25 , 30 , 45],
  "City":["Surat" , "Ahemedabad" , "Vadodara" , "Washington DC"]
}

print(student)
```

```python
df = pd.DataFrame(student)

print(df)
```

### Creation of DataFrame

```python
data = [
  ["Rahul" , 18],
  ["Rajesh" , 19],
  ["Rutul" , 20]
]

df = pd.DataFrame(data , columns=["Name" , "Age"])

print(df)
```

# Exploring DataFrame

```python
student = {
  "Name":["Rahul" , "Ritu" , "Rajesh" , "Dixit"],
  "Age":[20 , 25 , 30 , 45],
  "City":["Surat" , "Ahemedabad" , "Vadodara" , "Washington DC"]
}

df = pd.DataFrame(student)

print(df)
```

#### Display Last Rows

```python
print(df.tail(-3))
print(df.tail(3))
print(df.dtypes)
```

#### Shape of DataFrame

```python
print(df.shape)
print(df.columns)
```

#### Statistical Information

```python
print(df.describe())
print(df.info())
```

## Operations on DataFrame

#### Select single column

```python
print(df["Name"])
```

#### Select Multiple Column

```python
print(df[["Name" , "City"]])
```

#### Add New Column

```python
df["Grade"] = ["B" , "C" , "A" , "B"]

print(df)
```

#### Update Column

```python
df["Age"] = df["Age"] + 5

print(df)
```

#### Delete Column

```python
df.drop("City" , axis=1 , inplace=True)

print(df)
```

#### Filter Rows

```python
result = df[df["Age"] >= 30]

print(result)
```

#### Sorting Data

1. Ascending
2. Decending

```python
print(df.sort_values("Age"))
```

```python
print(df.sort_values("Age" , ascending=False))
```

```python
print(df["Age"].max())
```

```python
print(df["Age"].min())
```

```python
print(df["Age"].mean())
```

```python
print(df["Age"].sum())
```

# Pandas Series All Types with NaN Values 

#### what is NaN?

#### NaN [ Not a Number] respresent missing or undefined values in Pandas.

```python
import numpy as np
```

```python
s  = pd.Series([10 , 20 , np.nan , 40 , 50])

print(s)
```

#### 1. String Series with NaN

```python
s  = pd.Series(['Rahul' , 'Amit' , np.nan , 'Vivek' , 'Raj'])

print(s)
```

```python
s  = pd.Series([True , False , np.nan , True , False])

print(s)
```

```python
s  = pd.Series(["Amit" , False , np.nan , 20 , 30.5])

print(s)
```

#### Series With None

```python
s = pd.Series([10 , None , 30 , 40])

print(s)

print(s.isnull())

print(s.notnull())

print(s.isnull().sum())
```

#### Fill NaN Values

###### Using fillna()

```python
s = pd.Series([10 , np.nan , 30 , 40])

s = s.fillna(0)

print(s)
```

###### Using Dropna()

```python
s = pd.Series([10 , np.nan , 40])
print(s.dropna())
```
