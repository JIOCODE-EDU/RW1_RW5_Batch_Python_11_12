# Numpy Operation in Python with Jupyter Nootbook

#### import Numpy

```python
import numpy as np
```

#### Combining Arrays

#### Horizontal Stacks (hstack)

```python
a = np.array([1 , 2 , 3])
b = np.array([4 , 5 , 6])

result = np.hstack((a , b))

print(result)
```

#### Vertical Stack (vstack)

```python
a = np.array([1 , 2 , 3])
b = np.array([4 , 5 , 6])

result = np.vstack((a , b))

print(result)
```

## Splitting Arrays

```python
arr = np.array([1 , 2 , 3 , 4 , 5 , 6])

result = np.split(arr , 6)

print(result)
```

## Search Functions

#### Find Index

```python
arr = np.array([10 , 20 , 30 , 40 , 50])

index = np.where(arr == 50)

print(index)
```

#### Search Sorted Position

```python
arr = np.array([10 , 20 , 30 , 40 , 50])

position = np.searchsorted(arr , 30 , side='left')

print(position)
```

#### Sort Functions

```python
arr = np.array([50 , 30 , 300 , 20 , 10])

print(np.sort(arr))
```

## Aggregating & Statistical Functions

```python
arr = np.array([10 , 20 , 30 , 40 , 50])

print("sum:" , np.sum(arr))
print("mean:" , np.mean(arr))
print("Median:" , np.median(arr))
print("max:" , np.max(arr))
print("min:", np.min(arr))
```

#### Filter Arrays ( Boolean Indexing )

```python
arr = np.array([10 , 20 , 30 , 40 , 50])

result = arr[arr > 35]

print(result)
```

#### Multiple Condition

```python
arr = np.array([10 , 20 , 30 , 40 , 50])

result = arr[(arr > 20) & (arr < 50)]

print(result)
```

#### Masking Arrays

```python
arr = np.array([10 , 20 , 30 , 40 , 50])

mask = arr > 20

print(mask)

filterd = arr[mask]

print(filterd)
```

#### Reshaping Arrays

##### Convert array dimensions without changing data.

1D to 3D

```python
arr = np.array([10 , 20 , 30 , 40 , 50 , 60 , 70 , 80])

result = arr.reshape(2 , 2 , 2)

print(result)
```

1D to 2D

```python
arr = np.array([10 , 20 , 30 , 40])

result = arr.reshape(2 , 2)

print(result)
```

```python
arr = np.array([[[10,20],[30,40]],[[50,60],[70,80]]])

result = arr.reshape(8)

print(result)
```

#### Array Broadcasting

```python
a = np.array([[1] , [2] , [3] , [4]])

b = np.array([10 , 20 , 30])

result = a + b

print(result)
```
