##### Histogram Chart

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(10)
ages = np.random.randint(18 , 60 , 100)

plt.figure(figsize = (7 , 4))
plt.hist(ages , bins=10)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
month = ['Jan' , 'Feb' , 'Mar' , 'Apr' ,'May' , 'Jun' , 'July']
revenue = [120 , 160 , 100 , 170 , 220 , 240 , 200]
profit = [20 , 45 , 65 , 85 , 110 , 120 , 130]

plt.figure(figsize=(8 , 4))
plt.bar(month , revenue , label = "Revenue" , color=["red" , "blue" , "green" , "purple" , "orange" , "black" , "gray"])
plt.plot(month , profit , marker='o' , linewidth=2 , label="Profit")
plt.title("Revenue and Profit")
plt.xlabel("Month")
plt.ylabel("Amount")
plt.legend()
plt.show()
```

```python
months = ['Jan' , 'Feb' , 'Mar' , 'Apr' ,'May' , 'Jun' , 'July' , 'Aug' , 'Sep' , 'Oct' , 'Nov', 'Dec']

A = [20 , 25 , 30 , 35 , 40 , 45 , 50 , 38 , 50 , 55 , 60 , 64]
B = [20 , 45 , 30 , 35 , 50 , 45 , 50 , 78 , 50 , 55 , 20 , 64]
C = [20 , 25 , 70 , 35 , 40 , 45 , 40 , 38 , 40 , 55 , 40 , 34]

plt.figure(figsize=(10 , 5))

plt.stackplot(months , A , B , C , labels = ["Product A" , "Product B" , "Product C"] , colors = ["red" , "green" , "blue"])

plt.title("Monthly Sales of Products")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend(loc="center")
plt.show()
```

```python
months = ['Jan' , 'Feb' , 'Mar' , 'Apr' ,'May' , 'Jun']

HR = [20 , 25 , 30 , 35 , 40 , 45]
Marketing = [20 , 45 , 30 , 35 , 50 , 45]
Sales = [20 , 25 , 70 , 35 , 40 , 45]
IT = [20 , 25 , 50 , 45 , 10 , 65]

plt.figure(figsize=(10 , 5))

plt.stackplot(months , HR , Marketing , Sales , IT , labels = ["HR" , "Marketing" , "Sales" , "IT"] , colors = ["red" , "green" , "blue" , "orange"])

plt.title("Department Revuenue Distribution")
plt.xlabel("Months")
plt.ylabel("Revenue")
plt.legend(loc="best")
plt.savefig("revenue_department")
plt.show()
```
