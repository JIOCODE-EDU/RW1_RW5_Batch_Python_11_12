```python
import matplotlib.pyplot as plt
```

##### Create a Bar Chart

```python
products = ['product A', 'product B', 'product C', 'product D']
sales = [120 , 150 , 90 , 180]

plt.bar(products, sales)
plt.figure(figsize=(10, 4))
plt.title('Salses of Four Products')
plt.xlabel("Products")
plt.ylabel("Sales")
plt.show()
```

##### Create Line Chart

```python
months = ['January', 'February', 'March', 'April', 'May', 'June']
tempratures = [30, 32, 35, 40, 45, 50]

plt.figure(figsize=(8 , 4))
plt.plot(months , tempratures , marker="o")
plt.title('Montly Temperature')
plt.xlabel("Months")
plt.ylabel("Temperature")
plt.grid(True)
plt.show()
```

##### Create a Scatter Plot

```python
hours = [1 , 2 , 3 , 4 , 5 , 6 , 7 , 8 , 9 , 1 , 2 , 3 , 4 , 5 , 6 , 7 , 8 , 9]

scores = [35 , 40 , 45 , 50 , 60 , 65 , 88 ,42 , 58 , 58 , 88 , 62 , 42 , 65 , 60 , 47 , 58 , 22]


plt.figure(figsize=(6 , 4))
plt.scatter(hours , scores)
plt.title("Hours Studied vs Exam Scores")
plt.xlabel("Hours")
plt.ylabel("Exam Score")
plt.grid(True)
plt.show()
```

##### Create a Pie Chart

```python
fruits = ['Apple' , 'Banana' , 'Orange' , 'Mango' , 'Grapes']

quntity = [30 , 20 , 25 , 15 , 10]

plt.figure(figsize=(7 , 4))
plt.pie(quntity , labels=fruits , autopct='%1.1f%%' , startangle=90)
plt.title("Fruit Sales Distribution")
plt.show()
```
