# Pandas - Read and Manipulate one or multiple excel sheets


```python
import pandas as pd
```

```python
print("======= Read First Sheet =======")

df = pd.read_excel('student.xlsx')

print(df)
```

#### Read Specific Sheet

```python
sheet2 = pd.read_excel("student.xlsx" , sheet_name="Sheet2")

print(sheet2)
```

#### Read Sheet by Index

```python
print("======== Read Second Sheet ========")

sheet2 = pd.read_excel("student.xlsx", sheet_name= -1)

print(sheet2)
```

#### Read multiple sheet

```python
print("===== Read Multiple Sheet =======")

multiple_sheet = pd.read_excel("student.xlsx" , sheet_name=[0 , 1])

print(multiple_sheet)
```

#### Read All Sheets

```python
print("======== Read All Sheets ===========")

all_sheet = pd.read_excel("student.xlsx" , sheet_name=None)

for sheet_name , data in all_sheet.items():
  print("Sheet Name:" , sheet_name)
  print('----------------------------------------')
  print(data)
```

#### Select Data

```python
print("======== single column ===========")

print(df["Name"])

print("====== Multiple Column ========")

print(df[["Name" , "Marks"]])

print("====== First Row ==========")

print(df.iloc[1])
```

#### Missing Values

```python
print("======== Missing Values ========")

print(df.isnull().sum())

df["Marks"] = df["Marks"].fillna(df["Marks"].mean())

df["City"] = df["City"].fillna('Delhi')

print(df)
```

#### Remove Duplicate Records 

```python
print("\n Duplicate Rows :")
```

```python
df.duplicated().sum()

df = df.drop_duplicates()

print("\n After Duplicate")

print(df)
```

#### Change Data Types

```python
print(df.dtypes)

df["Age"] = df["Age"].astype(str)

print(df.dtypes)
```

#### Date & Time

```python
df["JoiningDate"] = pd.to_datetime(df["JoiningDate"])

df["Year"] = df["JoiningDate"].dt.year

df["Month"] = df["JoiningDate"].dt.month_name()

df["Day"] = df["JoiningDate"].dt.day_name()

print(df[["JoiningDate" , "Year" , "Month" , "Day"]])
```

#### Filter Data

```python
print("\n============ Marks > 80 ============")

result = df[df["Marks"] < 80]

city = df[df["City"] == "Surat"]

print(result)

print(city)
```

#### Sorting Data

```python
print("\n======== Ascending ===========")

result = df.sort_values("Marks")

print(result)


print("\n======== Descending ===========")

result = df.sort_values("Marks" , ascending=False)

print(result)
```

#### Add New Column

```python
df["Result"] = df["Marks"].apply(lambda x: "Pass" if x >= 35 else "Fail")

print("======  Result =======")

print(df)
```

#### Rename column

```python
df.rename(
  columns = {
    "Name" : "Student_Name"
  },
  inplace=True
)

print(df.columns)

print(df)
```

#### Delete Column

```python

df.drop('Age' , axis = 1 , inplace=True)

print(df)
```

#### Save Updated Data into Excel

```python
df.to_excel("student.xlsx" , index=False)

print("\n updated excel file saved!!")
```

# Pandas - Sorting , Filtering , Aggregation  , Merge and Join

```python
print("="*60)

import pandas as pd

student_data = {
  "ID":[101 , 102 , 103 , 104 , 105],
  "Name":["Vivek" , "Rahul" , "Neha" , "Priya" , "Zeel"],
  "Age":[28 , 25 , 21 , 23 , 50],
  "Marks":[89 , 78, 91 , None , 89],
  "Course":["Python" , "React" , "Python" , "Python" , "C++"],
  "City":["Surat" , "Ahmedabad" , "Vadodara" , None , "Rajkot"]
}

dt = pd.DataFrame(student_data)

print(dt)
```

### Merge Data

```python
print("\n Average Marks Course Wise")

print(dt.groupby("Course")["Marks"].mean())


print("\n Maximum Marks Course Wise")

print(dt.groupby("Course")["Marks"].max())

print("\n Minimum Marks Course Wise")

print(dt.groupby("Course")["Marks"].min())

print("\n Count Course wise student")

print(dt.groupby("Course")["ID"].count())
```

```python
fees = {
  "ID":[101 , 102 , 103 , 104 , 105],
  "Fees":[25000 , 22000 , 50000 , 150000 , 20000]
}

fee_dt = pd.DataFrame(fees)

print(fee_dt)
```

#### Merge

```python
print("Inner Merge")

inner = pd.merge(dt , fee_dt , on="ID" , how="right")

print(inner)
```
